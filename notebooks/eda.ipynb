```python
import pandas as pd
```

```python
media_variables = pd.read_excel("../data/Dataset UK L'Oreal Paris Haircare - HEC Training.xlsx", sheet_name = "A&P Variables")
target_variables = pd.read_excel("../data/Dataset UK L'Oreal Paris Haircare - HEC Training.xlsx")
control_variables = pd.read_excel("../data/Dataset UK L'Oreal Paris Haircare - HEC Training.xlsx", sheet_name = "Commercial Variables")
```

We need to build one big dataframe with the following columns :
- date
- target_variable (revenue or sold units)
- media_impressions and media_spend for each media
- control_variable

```python
# Let's join those 2

controls_target_variables = control_variables.merge(target_variables, on=["Starting Week", "Year"], how="left", validate="one_to_one").sort_values("Starting Week", ascending = True)
controls_target_variables.rename(columns= {"Starting Week": "Starting week"}, inplace = True)
```

```python
# Let's isolate tv, advocacy media and shopper_experience

media_tv = media_variables[media_variables["growth_driver_l4"]=="traditional_tv"]
media_engagements = media_variables[media_variables["metric"]=="engagements"]
media_shopper_experience = media_variables[media_variables["growth_driver_l4"]=="testers_and_merchandising"]
```

```python
media_variables = media_variables[media_variables["metric"]=="impressions"]
```

```python
media_impressions_l4 = media_variables.groupby(["Starting week", "growth_driver_l4"]).agg(
    spend = ("investment (in pound)", "sum"),
    impressions=("execution", "sum")
).reset_index()
```

```python
media_impressions_l5 = media_variables.groupby(["Starting week", "growth_driver_l5"]).agg(
    spend = ("investment (in pound)", "sum"),
    impressions=("execution", "sum")
).reset_index()
```

```python
media_impressions_l4.head()
```

```python
media_impressions_l4['growth_driver_l4'] = media_impressions_l4['growth_driver_l4'].str.lower()

l4_df = (
    media_impressions_l4
    .pivot_table(
        index='Starting week',
        columns='growth_driver_l4',
        values=['spend', 'impressions'],
        aggfunc='sum',
        fill_value=0
    )
)
l4_df.columns = [
    f"{metric}_{media}"
    for metric, media in l4_df.columns
]

# optional: reset index
l4_df = l4_df.reset_index().sort_values("Starting week", ascending = True)
```

```python
media_tv = media_tv.groupby(["Starting week", "growth_driver_l4"]).agg(
    spend=("investment (in pound)", "sum"),
    grp = ("execution", "mean")
)
```

```python
l4_grp_df = (
    media_tv
    .pivot_table(
        index='Starting week',
        columns='growth_driver_l4',
        values=['spend', 'grp'],
        aggfunc='sum',
        fill_value=0
    )
)
l4_grp_df.columns = [
    f"{metric}_{media}"
    for metric, media in l4_grp_df.columns
]

l4_grp_df = l4_grp_df.reset_index().sort_values("Starting week", ascending = True)
```

```python
media_engagements = media_engagements.groupby(["Starting week", "growth_driver_l4"]).agg(
    spend=("investment (in pound)", "sum"),
    engagement = ("execution", "sum")
)
l4_engagement_df = (
    media_engagements
    .pivot_table(
        index='Starting week',
        columns='growth_driver_l4',
        values=['spend', 'engagement'],
        aggfunc='sum',
        fill_value=0
    )
)
l4_engagement_df.columns = [
    f"{metric}_{media}"
    for metric, media in l4_engagement_df.columns
]

l4_engagement_df = l4_engagement_df.reset_index().sort_values("Starting week", ascending = True)
```

```python
# let's merge everything

final_df = pd.concat([controls_target_variables, l4_df, l4_grp_df, l4_engagement_df], axis = 1)
```
